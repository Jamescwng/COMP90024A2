```python
rows = db.view('_design/CrimeInfo/_view/GeoData')
```

```python
data = [row['value'] for row in rows]
random.shuffle(data)
##df = pd.DataFrame(data)
##df.head()
```

```python
MELB_COORDINATES = (-37.810612, 144.963954)
```

```python
_map = folium.Map(location=MELB_COORDINATES, zoom_start=13)

HeatMap(data[:1500]).add_to(_map)

_map
```

```python
#_map = folium.Map(location=MELB_COORDINATES, zoom_start=12)

mCluster = MarkerCluster(
    name="Crimes",
    overlay=True,
    control=True
)

for each in data[:1500]:
    
    html = '''<br>Latitude: {}<br>Longitude:{}<br>'''.format(each[0], each[1])
    iframe = folium.IFrame(html, width=200, height=200)
    popup = folium.Popup(iframe, max_width=200)
    marker = folium.Marker(location=[each[0], each[1]],popup=popup)
    mCluster.add_child(marker)

folium.LayerControl().add_to(_map)

_map
```
